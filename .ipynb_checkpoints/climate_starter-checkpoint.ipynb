# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Use Inspector to print the column names and types
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c["type"])
```

```python
# Use Inspector to print the column names and types
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c["type"])
```

```python
# reflect an existing database into a new model
Base = automap_base()
Base.prepare(engine, reflect = True)
Base.classes.keys()  # does the same as inpector.get_table_names()
# reflect the tables
Station = Base.classes.station
Measurement = Base.classes.measurement
```

```python
# We can view all of the classes that automap found
session = Session(engine)
result = session.query(Measurement.date, Measurement.prcp).all()
result
```

```python
# Save references to each table
engine.execute('SELECT * FROM Measurement LIMIT 10').fetchall()
```

```python
# Create our session (link) from Python to the DB

engine.execute('SELECT * FROM Station').fetchall()
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
# Calculate the date 1 year ago from the last data point in the database
latest_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
latest_date[0]
```

```python
latest_date = dt.date(int(latest_date[0].split("-",1)[0]),int(latest_date[0].split("-",2)[1]),int(latest_date[0].split("-",2)[2]))
date_last_year = latest_date - dt.timedelta(days = 365)
print(date_last_year)
```

```python
# Perform a query to retrieve the data and precipitation scores
results = session.query(Measurement.date, Measurement.prcp).\
          filter(Measurement.date >= date_last_year).all()
# Save the query results as a Pandas DataFrame and set the index to the date column
prcp_results = pd.DataFrame(results).set_index("date")
#prcp_result = [result[1] for result in results]
prcp_results
# Sort the dataframe by date
sorted_prcp_results = prcp_results.sort_values("date", ascending = True).groupby("date").max() 
# Use Pandas Plotting with Matplotlib to plot the data
x_axis = np.arange(len(sorted_prcp_results["prcp"]))
y_axis = sorted_prcp_results["prcp"]
plt.bar(x_axis, y_axis, width = 5 ,alpha = 0.8, align = 'center')
## Label the tick locations
date_label = []
tick_locations = np.arange(0, len(x_axis), 35)
date_tolist = sorted_prcp_results.index.tolist()
loc_value = [location for location in tick_locations]
for num in loc_value:
    date_label.append(date_tolist[num])
##
plt.xticks(tick_locations, date_label, rotation = "vertical")
plt.xlim(-2,len(x_axis)+2)
plt.ylim(0, max(y_axis)+0.35)
plt.xlabel("Date")
plt.ylabel("Inches")
plt.show()
#plt.savefig("../Images/precipitation.png")
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data in the last year
prcp_count = prcp_results.count()
prcp_mean = prcp_results.mean()
prcp_std = prcp_results.std()
prcp_min = prcp_results.min()
#25%
quartiles = prcp_results["prcp"].quantile([0.25,0.5,0.75])
lowerq = quartiles[0.25]
lowerq
#50%
midq = quartiles[0.5]
midq
#75%
upperq = quartiles[0.75]
upperq
prcp_max = prcp_results.max()

prcp_stats = {"precipitation" : [prcp_count[0], prcp_mean[0],prcp_std[0], prcp_min[0], lowerq, midq, upperq, prcp_max[0]]}
pd.DataFrame(prcp_stats).rename(index = {0:"count",1: "mean", 2:"std", 3:"min", 4:"25%", 5:"50%", 6:"75%", 7:"max"})
```

```python
# Design a query to show how many stations are available in this dataset?
stations = session.query(Measurement.station).group_by(Measurement.station).all()
station_count = len(stations)
station_count
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
active_stations = session.query(Measurement.station, func.count(Measurement.tobs)).\
                  group_by(Measurement.station).\
                  order_by(func.count(Measurement.tobs).desc()).all()
most_active_station = active_stations[0][0]
most_active_station
```

```python
# List the stations and the counts in descending order.
active_stations
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
sel = [Measurement.station,\
       func.min(Measurement.tobs),\
       func.max(Measurement.tobs),\
       func.avg(Measurement.tobs)]
temp_data = session.query(*sel).\
            filter(Measurement.station == most_active_station).\
            group_by(Measurement.station).all()
temp_data
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temp_last_year = session.query(Measurement.station, Measurement.tobs).\
                 filter(Measurement.date >= date_last_year).\
                 filter(Measurement.station == most_active_station).all()
temp_last_year
```

```python
temp_df = pd.DataFrame(temp_last_year, columns = ["station", "tobs"])
temp_df.plot.hist(bins = 12)
plt.xlabel("Temperature")
plt.tight_layout()
plt.show()
#plt.savefig("../Images/station-histogram.png")
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
trip_temp_df = pd.DataFrame({"tmin":[trip_temp[0][0]],"tavg":[trip_temp[0][1]],"tmax":[trip_temp[0][2]]})

error = trip_temp_df["tmin"].sub(trip_temp_df["tmax"])
trip_temp_df.plot.bar(y="tavg", yerr = error)
plt.xlim(-1,1)
plt.title("Trip Avg Temp")
plt.tight_layout()
plt.show()
plt.savefig("../Images/temperature.png")
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
